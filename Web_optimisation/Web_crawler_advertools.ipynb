# **Before we start**
In this tutorial, I will show you how to use a python package called advertools to crawl a website easily and do some technical SEO or some other analysis for web optimisation.

Advertools provides online marketing productivity and analysis tools in Pyrthon. This oython package can help us easily crawl a website and return back the results in an organised way (in dataframe). This will save us lots of time cleaning the data and you can focus on the analysis of the data. You can read more about Advertools through this [link](https://advertools.readthedocs.io/en/master/readme.html).

There are various tools available in Advertools package. In this tutorial, we will focus on using Advertools to build a SEO spider/crawler.

And then import the medule called `crawl` to enable us crawl the website.

```python
import pandas as pd
from advertools import crawl
```

In this tutorial, we are crawling the world-renowned website: p-tech.org.uk :)

1) The first line has crawled the “p-tech.org.uk”. We have crawled the website and output the data in “jl” format. “follow_links=True” attribute is for making crawlers following the URLs on a web page.

2) We have created a variable that is “ptech”, we have used the “ptech.jl” output via “lines=True” attribution so that our crawled data can be seen in a data frame.

3) We have called our variable’s data using ptech.head()

```python
crawl('https://www.p-tech.org.uk/', 'ptech.jl', follow_links=True)
ptech = pd.read_json('ptech.jl', lines=True)
ptech.head(5)
```

That's it! We have successfully crawled p-tech website and this is what the module has done while we run the previous code:


1) Check the site's robots.txt file and get the crawl rules, which means that your crawl will be affected by these rules and the user agent you are using. In some cases, we need to change the user agents to control this.

2) Starting with the provided URL(s) go through all links and parse pages.

3) For each URL extract the most important SEO elements.

4) Save them to my_output_file.jl (`jl`-Jsonlines is the supported output format because of its flexibility in allowing different values for different scraped pages, and appending indepentent items to the output files.)

5) read the json file and then transform it into the dataframe. The column headers of the output file (once you import it as a DataFrame) would be the names of the elements (title, h1, h2, etc.).



We can get all the columns and check what SEO elements are extracted. There are 68 columns and you can check the further details of these elements through this [link](https://advertools.readthedocs.io/en/master/advertools.spider.html). Basically, this tool has helped us crawl all the urls from the website and also return some content information such as title, meta-description,images and etc. and also some information of the crawl such as crawl-time, status code, redirect and etc. Thus, we can use these data to analyse the website and understand the issues with the website.

```python
ptech.columns.tolist()
```

Next, I am showing you some basic analysis you can do with this data.

1) You can check the status code and understand whether all the pages are working or whether there is any broken link. The data is in the "status" column.

```python
ptech['status'].value_counts()
```

By running the above code, we can find that the majority of the pages with status code 200, which means these pages can be accessed successfully. However, there are 8 pages coming back with 404 error, that means these pages are broken or the content unavailable. We can further check which pages are 404 pages.

```python
ptech[ptech['status']==404]['url']
```

We have found all 8 pages with 404 error and for the business we need to further check the issues and optimise pages for better customer experience. There are different practices for 404 errors. We can remove them of course or we can set the redirect or we can put some further useful information in this page if we still want to keep this 404 page (for example, this 404 page is the unavailable product and this product is bringing us traffic then we can keep it but provide some further recommendations in this page for users who visit this 404 page.)

2) We can also check whether there are missing page titles or meta descriptions in the website. As we know, we need to make sure we have set the title and meta description for good practice of SEO. By setting some keywords properly, this will allow the webiste easier to be found in the search engine.

```python
ptech['title'].isna().value_counts() # check the nulls in title column to understand whether we have missing titles
```

Good news is that we don't have missing titles in p-tech website.

```python
ptech['meta_desc'].isna().value_counts()  #check the nulls in meta_desc column to understand whether we have missing metadescription
```

But unfortunately we have some pages without meta descriptions and this is the opportunity for improvements.

Not just checking missing titles, we can also check the length of titles. For good practice of SEO, we cannot have too long or too short title. It is quite common to suggest titles are no longer than 60 characters. Here we check how many titles are longer than 60 characters.

```python
(ptech['title'].str.len() > 60).value_counts()
```

There are 37 page titles longer than 60 characters, then we can think we can optimise these titles better for SEO purpose. But there is no such rule we have to keep the titles within 60 characters. Just longer than 60 charaters it can be cut while showing the search result pages and that might impact the customer experience.

```python
ptech[ptech['title'].str.len() > 60].filter(['url','title'],axis=1)
```

3) We can also check the frequently used words in title or meta description. And get some ideas about the keyword used in this website. We are using simple word count to do this. Of course we can use more comprehensive text analaysis (NLP techniques) but this would be another topic we will cover later in the third week.

```python
import advertools as adv
title_words = str(ptech['title'].tolist()).replace("-","") # there is "-" separator in the title and we just remove it so it won't apprear in the list
adv.word_frequency(title_words, rm_words=adv.stopwords['english']).head(20)
```

```python
meta_desc_words = str(ptech['meta_desc'].dropna().tolist()) # here we drop the urls without meta descriptions using .dropna() function
adv.word_frequency(meta_desc_words, rm_words=adv.stopwords['english']).head(20)
```
