```python
# parse the log file and save as .parquet format
adv.logs_to_df(log_file='arrowsmithmarketing.log',
               output_file='arrowsmithmarketing.parquet',
               errors_file='arrowsmithmarketing.txt',
               log_format='combined')
```

```python
# Read the parquet file into a pandas DataFrame, and convert the datetime column into a datetime object.
logs_df = pd.read_parquet('arrowsmithmarketing.parquet')
logs_df['datetime'] = pd.to_datetime(logs_df['datetime'],
                                     format='%d/%b/%Y:%H:%M:%S %z')

logs_df
```

```python
# Split the request URLs into their components.
request_url_df = adv.url_to_df(logs_df['request'])
request_url_df = request_url_df.add_prefix('request_')
request_url_df.head(10)
```

```python
# Split the referer URLs into their components.
referer_url_df = adv.url_to_df(logs_df['referer'])
referer_url_df = referer_url_df.add_prefix('referer_')
referer_url_df.head(10)
```

```python
!pip install ua_parser
from ua_parser import user_agent_parser
pd.options.display.max_columns = None
```

```python
# parse the user_agent column
ua_df = pd.json_normalize([user_agent_parser.Parse(ua) for ua in logs_df['user_agent']])
ua_df.columns = 'ua_' + ua_df.columns.str.replace('user_agent\.', '', regex=True)
ua_df.head(10)
```

```python
# merge all the daframe as one
(pd.concat([logs_df, request_url_df, referer_url_df, ua_df], axis=1)
.to_parquet('logs_final.parquet', index=False, version='2.4'))
```

```python
# we can read the new file in dataframe and can start analysis
new_logs_df = pd.read_parquet('logs_final.parquet')
new_logs_df.head()
```
